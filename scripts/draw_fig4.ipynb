```python
import matplotlib.pyplot as plt

# Sample data (you'll need to replace this with your actual data)
x_values = [10, 15, 20, 25, 30]
a_thaliana_series1 = [0.937033832073211, 0.953273594379425,0.96130633354187, 0.965854465961456, 0.968749165534973]
b_vulgaris_series1 = [0.963338911533355, 0.969398558139801, 0.972083628177642, 0.973664760589599, 0.974713504314422]

# Creating the figure and subplots
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(12, 4), sharey=True)

# Plotting for A. thaliana
ax1.plot(x_values, a_thaliana_series1, marker='o', label='correlation')
ax1.set_title('A. thaliana')
ax1.set_xticks(x_values)
ax1.set_xticklabels([f"{x}x" for x in x_values])
ax1.set_ylim(0.85, 1.0)
ax1.grid(color='gray', linestyle='-', linewidth=0.5)  # 添加灰色网格线

# Plotting for B. vulgaris
ax2.plot(x_values, b_vulgaris_series1, marker='o', label='correlation')
ax2.set_title('O. sativa')
ax2.set_xticks(x_values)
ax2.set_xticklabels([f"{x}x" for x in x_values])
ax2.grid(color='gray', linestyle='-', linewidth=0.5)  # 添加灰色网格线

# Adding common labels
fig.text(0.5, 0.0, 'Sequencing Coverage Depth', ha='center')
fig.text(0.07, 0.5, 'Correlation Coefficient', va='center', rotation='vertical')

# Displaying the legend in one of the plots
ax1.legend(loc='lower right')
ax2.legend(loc='lower right')

# plt.tight_layout()
plt.savefig("fig4_3", dpi=400, bbox_inches="tight")
plt.show()
```

```python
import matplotlib.pyplot as plt

# 样例数据
x_values = [10, 15, 20, 25, 30]  # 覆盖深度
correlation_coefficients = [0.937033832073211, 0.953273594379425,0.96130633354187, 0.965854465961456, 0.968749165534973]  # 相关系数

# 创建图形
fig, ax = plt.subplots(figsize=(6, 4))

# 绘制数据线条
ax.plot(x_values, correlation_coefficients, marker='o', label='correlation')

# 设置标题和标签
ax.set_title('A. thaliana')
ax.set_xlabel('Coverage Depth')
ax.set_ylabel('Correlation Coefficient')

# 设置x轴的刻度标签
ax.set_xticks(x_values)
ax.set_xticklabels([f"{x}x" for x in x_values])

# 设置y轴的范围
ax.set_ylim(0.86, 1.0)

# 添加网格线
ax.grid(True, color='grey', linestyle='-', linewidth=0.5)

# 添加图例
ax.legend(loc='lower right')

# 显示图表
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Generate random data to simulate the heatmap values
# Replace this with your actual data if available
# data = np.random.rand(20, 20)
data1 = np.random.rand(100, 1)
# data1[:50] += 1
# data1[50:] -= 1
data2 = np.random.rand(1, 100) * 0.01 + data1.reshape(1, 100)
data = np.matmul(data1, data2)

# Create the heatmap
plt.figure(figsize=(6, 6))
sns.heatmap(data, cmap='Set3', cbar=False, square=True)

# Customize labels and add annotation text
plt.xlabel("WGBS")
plt.ylabel("")
plt.xticks(ticks=[0, 25, 50, 75, 100], labels=["0.00", "0.25", "0.50", "0.75", "1.00"])
plt.yticks(ticks=[0, 25, 50, 75, 100], labels=["1.00", "0.75", "0.50", "0.25", "0.00"])

# Add annotations for the values of N and r
plt.text(0.5, 1.05, r"$N=48.0\ M \quad r=0.918$", ha='center', va='center', transform=plt.gca().transAxes)

# Display the plot
plt.tight_layout()
plt.show()
```
